```python
import anthropic
import dotenv
print("Both installed!")

# Load env variables
from dotenv import load_dotenv

load_dotenv()
print("Env variables loaded!")

# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
print("Client created!")
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chatbot(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text

#---------------------------------------------------------------------------------------------------------
messages = []

import textwrap
system_prompt="Respond in plain text only. Do not use markdown formatting. "
user_input = "You are a Warner Bros. executive. Generate a one sentence tv series idea."

# Prompt for User Input for messages
#while True:
#    user_input = input("Enter your question (or 'q' to quit): ")
#    print (">", user_input)
#    if user_input.lower() == ("q"):
#        print("Goodbye!")
#        break

add_user_message(messages, user_input)
response = chatbot(messages, system=system_prompt, temperature=1.0)
#add_assistant_message(messages, response)
print(textwrap.fill(response, width=80))


```
