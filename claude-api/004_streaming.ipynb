```python
import anthropic
import dotenv
print("Both installed!")

# Load env variables
from dotenv import load_dotenv

load_dotenv()
print("Env variables loaded!")

# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
print("Client created!")
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

#---------------------------------------------------------------------------------------------------------
messages = []
user_input = "Write a haiku."

add_user_message(messages, user_input)
stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)
for event in stream:
    print(event)
```

```python
messages = []
user_input = "Generate a very short event bridge rule in JSON format."
add_user_message(messages, user_input)

def chatbot(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text

answer = chatbot(messages)
print(answer)
```

```python
# Structured Data Exercise
messages = []
user_input = "Generate three different sample AWS CLI commands. Each should be very short."
add_user_message(messages, user_input)

def chatbot(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        system="Do not respond with markdown or any explanation.",
        messages=messages,
    )
    return message.content[0].text.strip()

answer = chatbot(messages)
print(answer.replace('\n\n', ' || '))
```
