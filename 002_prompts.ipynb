```python
import anthropic
import dotenv
print("Both installed!")

# Load env variables
from dotenv import load_dotenv

load_dotenv()
print("Env variables loaded!")

# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
print("Client created!")
```

```python
# System Prompts
# Math Tutor Exercise

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

# New function for math tutor, with updated system instructions
def chatbot(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text


#---------------------------------------------------------------------------------------------------------

system_prompt="Respond in plain text only. Do not use markdown formatting. " \
              "Responses should be step by step and detailed. Do not solve the problem, just give hints and guidance." \
              "Responses should be in the context of a math tutor helping a student understand how to solve a problem." \
              "Responses should be printed line by line, with a pause between each line, to simulate a tutor writing on a chalkboard."


#Initial List of messages
messages = []

# Prompt for User Input for messages
while True:
    user_input = input("Enter your question (or 'q' to quit): ")
    print (">", user_input)
    if user_input.lower() == ("q"):
        print("Goodbye!")
        break

    add_user_message(messages, user_input)

    # Get response
    response = chatbot(messages, system=system_prompt)
    add_assistant_message(messages, response)
    print(response)
```

```python
# System Prompt Exercise
#Initial List of messages
messages = []

question = "Write a Python function that checks a string for duplicate characters and returns a list of the duplicate characters."
system_prompt= "Respond in plain text only. Do not use markdown formatting." \
               "Responses should be concise and to the point, providing only the necessary information to answer the question without any additional explanation or context."

add_user_message(messages, question)
response = chatbot(messages, system=system_prompt)
print(response)
```
