```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
print("Client created!")
```

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        system="Respond in plain text only. Do not use markdown formatting. ",
        messages=messages,
    )
    return message.content[0].text


#message = client.messages.create(
#    model=model,
#    max_tokens=1000,
#    messages=[
#        {
#            "role": "assistant",
#            "content": message.content[0].text
#        },
#        {
#            "role": "user",
#            "content": "Give another sentence to the previous answer you provided."
#        }
#    ]
#)
```

```python
# Starting list of messages
messages = []

# Initial User message
add_user_message(messages, "What is the meaning of life? Answer in one sentence.")

print("=" * 20)
print("First MESSAGES LIST")
print("=" * 20)
for m in messages:
    print(f"[{m['role'].upper()}]")
    print(m['content'])
    print()


# Pass first message into the "chat" function
answer = chat(messages)

import textwrap
print("=" * 20)
print("First Answer")
print("=" * 20)
print(textwrap.fill(answer, width=100))
print()

# Pass the answer output back as an assistant message, and add a follow up user message
add_assistant_message(messages, answer)
add_user_message(messages, "Give another sentence to the previous answer you provided.")

print("=" * 20)
print("Appended MESSAGES LIST")
print("=" * 20)
for m in messages:
    print(f"[{m['role'].upper()}]")
    print(m['content'])
    print()


# Pass the updated list of messages back into the "chat" function
answer = chat(messages)

print("=" * 20)
print("Follow-up Answer")
print("=" * 20)
print(textwrap.fill(answer, width=100))
print()
```

```python
# Chatbox Exercise

#Initial List of messages
messages = []

# Prompt for User Input for messages
while True:
    user_input = input("Enter your message (or 'q' to quit): ")
    print (">", user_input)
    if user_input.lower() == ("q"):
        print("Goodbye!")
        break

    add_user_message(messages, user_input)

    # Get response
    response = chat(messages)
    add_assistant_message(messages, response)
    print(textwrap.fill(response, width=100))
    print()
```
